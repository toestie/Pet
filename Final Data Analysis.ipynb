```python
#import dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
#Are particular breeds being adopted? (import data)
breedinpath = "output/breedin_count.csv"
adoptoutpath = "output/breedout_count.csv"
breedin_df = pd.read_csv(breedinpath)
adoptout_df = pd.read_csv(adoptoutpath)
```

```python
#Are particular breeds being adopted? (Intake Graph)
breedin_x = breedin_df["Count"][0:10]
breedin_labels = breedin_df["Breed"][0:10]
plt.barh(breedin_labels, breedin_x)
plt.title("Most Common Breeds (intake)")
plt.xlabel("Count")
plt.savefig("images/breedin.png")
plt.show()
```

```python
#Are particular breeds being adopted? (Adopt Graph)
adoptout_x = adoptout_df["Count"][0:10]
adoptout_labels = adoptout_df["Breed"][0:10]
plt.barh(adoptout_labels, adoptout_x)
plt.title("Most Common Breeds (adopted)")
plt.xlabel("Count")
plt.savefig("images/adoptout.png")
plt.show()
```

```python
# The differences between the intaken and adopted are not significant
(t_stat, p) = stats.ttest_ind(breedin_x, adoptout_x, equal_var=False)
p
```

```python
#What are the trends in adoption and euthanasia? (import data)
euthanasiaDogpath = "output/euthanasiaDog.csv"
euthanasiaCatpath = "output/euthanasiaCat.csv"
adoptionDogpath = "output/adoptionDog.csv"
adoptionCatpath = "output/adoptionCat.csv"

euthanasiaDog_df = pd.read_csv(euthanasiaDogpath)
euthanasiaCat_df = pd.read_csv(euthanasiaCatpath)
adoptionDog_df = pd.read_csv(adoptionDogpath)
adoptionCat_df = pd.read_csv(adoptionCatpath)
```

```python
#What are the trends in adoption and euthanasia? (adoption graph)
myaxis=["2013-12","2014-6","2014-12","2015-6","2015-12","2016-6","2016-12","2017-6","2-17-12"]
plt.plot(adoptionDog_df["monthyear"], adoptionDog_df["Count"], label="Dog")
plt.plot(adoptionCat_df["monthyear"], adoptionCat_df["Count"], label="Cat")
plt.ylabel("Count")
plt.xlabel("Year-Month")
plt.title("Adoptions from 2013 - 2017")
plt.grid()
plt.legend()
plt.xticks([0,6,12,18,24,30,36,42,48],myaxis,rotation=90)
plt.ylim(0,600)
plt.savefig("images/adoption.png")
plt.show
```

```python
#What are the trends in adoption and euthanasia? (euthanasia graph)
myaxis=["2013-12","2014-6","2014-12","2015-6","2015-12","2016-6","2016-12","2017-6","2-17-12"]
plt.plot(euthanasiaDog_df["monthyear"], euthanasiaDog_df["Count"], label="Dog")
plt.plot(euthanasiaCat_df["monthyear"], euthanasiaCat_df["Count"], label="Cat")
plt.ylabel("Count")
plt.xlabel("Year-Month")
plt.title("Euthanasias from 2013 - 2017")
plt.grid()
plt.legend()
plt.xticks([0,6,12,18,24,30,36,42,48],myaxis,rotation=90)
plt.ylim(0,90)
plt.savefig("images/euthanasia.png")
plt.show
```

```python
#Does Austin prioritize keeping the animal alive? (import data)
outcomepath = "output/outcome_count.csv"
outcome_df = pd.read_csv(outcomepath)
```

```python
#Does Austin prioritize keeping the animal alive? (graph all outcomes)
plt.bar(outcome_df["Outcome"], outcome_df["Count"])
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.title("All Outcomes from 2013 - 2017")
plt.savefig("images/outcomes.png")
plt.show()
```

```python
#Does Austin prioritize keeping the animal alive? (graph euthanasia vs adoption)
versus_df = outcome_df.loc[outcome_df['Outcome'].isin(['Adoption','Euthanasia'])]
plt.bar(versus_df["Outcome"], versus_df["Count"])
plt.ylabel("Count")
plt.title("Adoptions vs. Euthanasias from 2013 - 2017")
plt.savefig("images/adopteuthanize.png")
plt.show()
```
