```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("austin-animal-center-statistics-oct-1-2013-to-dec-7-2017-QueryResult (25).csv")
df=df.loc[df['monthyear'] != '2017-12']


df
```

```python
adoption=df[df.outcome_type=="Adoption"].groupby('monthyear').count()
adoptionDog=df[df.outcome_type=="Adoption"][df.animal_type=="Dog"].groupby('monthyear').count()
adoptionCat=df[df.outcome_type=="Adoption"][df.animal_type=="Cat"].groupby('monthyear').count()
adoptionOther=df[df.outcome_type=="Adoption"][df.animal_type=="Other"].groupby('monthyear').count()
adoptionCat
```

```python
myaxis=["2013-12","2014-6","2014-12","2015-6","2015-12","2016-6","2016-12","2017-6","2-17-12"]
plt.plot(adoption.index, adoption.animal_id,label="total")
plt.plot(adoptionDog.index, adoptionDog.animal_id, label="Dog")
plt.plot(adoptionCat.index, adoptionCat.animal_id, label="Cat")
plt.grid()
plt.legend()
plt.xticks([0,6,12,18,24,30,36,42,48],myaxis,rotation=90)
plt.ylim(0,1000)
plt.show
```

```python
euthanasia=df[df.outcome_type=="Euthanasia"].groupby('monthyear').count()
euthansiaDog=df[df.outcome_type=="Euthanasia"][df.animal_type=='Dog'].groupby('monthyear').count()
euthansiaCat=df[df.outcome_type=="Euthanasia"][df.animal_type=='Cat'].groupby('monthyear').count()
euthansiaOther=df[df.outcome_type=="Euthanasia"][df.animal_type=='Other'].groupby('monthyear').count()

```

```python
plt.plot(euthanasia.index, euthanasia.animal_id,label="total")
plt.plot(euthansiaOther.index, euthansiaOther.animal_id, label="Other")

plt.grid()
plt.legend()
plt.xticks([0,6,12,18,24,30,36,42,48],myaxis,rotation=90)
plt.ylim(0,300)
plt.show
```

```python
plt.plot(euthansiaDog.index, euthansiaDog.animal_id, label="Dog")
plt.plot(euthansiaCat.index, euthansiaCat.animal_id, label="Cat")
plt.grid()
plt.legend()
plt.xticks([0,6,12,18,24,30,36,42,48],myaxis,rotation=90)
plt.ylim(0,100)
plt.show
```

```python
euthanasia.plot(y="animal_id",use_index=True)
euthansiaDog.plot(y="animal_id",use_index=True)
euthansiaCat.plot(y="animal_id",use_index=True)
euthansiaOther.plot(y="animal_id",use_index=True)
```
